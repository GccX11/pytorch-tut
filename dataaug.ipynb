# load data

```python
my_transforms = transforms.Compose([
    transforms.ToPILImage(), # need to convert to this format
    transforms.Resize((256,256)), # resize to 256x256
    transforms.ColorJitter(brightness=0.5), # change brightness with a probability of 0.5
    transforms.RandomHorizontalFlip(p=0.5), # with probability 0.5, flip the image
    transforms.RandomCrop((224,224)), # crop the image to 224x224 at a random location
    transforms.RandomRotation(degrees=45), # rotate the image by 45 degrees
    #transforms.RandomGrayscale(p=0.2), # convert the image to grayscale with a probability of 0.2
    transforms.ToTensor() # convert back to tensor
    #transforms.Normalize(mean=[],
    #                     std=[]) # normalize the image
])

dataset = CatsAndDogsDataset(csv_file='labels.csv', 
                             root_dir='dataset/PetImages', 
                             transform=my_transforms)
```

```python
for i, (im, label) in enumerate(dataset):
    print(im.shape)
    plt.imshow(torch.mean(im, axis=0).detach().cpu().numpy(), cmap='gray')
    plt.show()
    if i > 2:
        break
```
